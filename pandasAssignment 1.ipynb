#4 removing space from column names

```python

superstoreData.columns=superstoreData.columns.str.replace(" ","")
print(superstoreData.columns.to_list())

#various ways to rename a column would be :- 
print('1) using map')
rmSpace1=list(map(lambda item:item.replace(' ','_'),superstoreData.columns))
print(rmSpace1)

print("2) using list comprehension")
rmSpace2=[i.replace(' ','_') for i in superstoreData.columns]
print(rmSpace2)

```
